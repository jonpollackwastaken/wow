```python
def pretty_print(s):
    caret = 0
    tab_count = 0
    prev=''
    uncaught = []
    for c in s:
        if c=='^' or prev =='^':
            cmd = prev+c

            if   cmd == '^T': tab_count+=1
            elif cmd == '^S': print('\n'+'\t'*tab_count, end='')
            elif cmd == '^t': tab_count-=1
            elif cmd == '^b': print('=UNSET', end='')
            elif cmd == '^N': print('=', end='')

            # uncaught commands
            elif prev=='^': uncaught+=[cmd]
        else: 
            print(c, end='')
        prev = c
    print('#'*10,'\n',uncaught)

parse(s)
```

```python
## setup function to take dict so that unnecessary parameters can be ignored

def dispatcher(fn, delim, item_list):
    return delim.join([fn(**item) for item in item_list])

def multi_auction_operation(item_list):
    return dispatcher(auction_operation, '^S^t^S', item_list)

def multi_shopping_operation(item_list):
    return dispatcher(shopping_operation, '^t^t^S', item_list)

def multi_group_operation_map(item_list):
    return dispatcher(group_operation_map, '^t^t^S', item_list)

def multi_group(item_list):
    return dispatcher(group, ',', item_list)
```
